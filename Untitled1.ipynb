```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
alpha = 0.33  # Capital share of output
beta = 0.99  # Discount factor
z_star = 0.3 #Technology Shock
rho = 0.8  # Persistence of technology shock
sigma = 0.1  # Standard deviation of technology shock
phi = 0.5  # Calvo parameter
theta = 2.5  # Elasticity of substitution
pi_star = 0.02  # Target inflation rate
delta = 0.1  # Depreciation rate
pi = 1
s = 0.6  # Share of firms that can adjust prices
gamma = 0.5  # Habit formation parameter
```

```python
# Steady-state values
k_star = ((1 - beta * (1 - delta)) / (alpha * beta)) ** (1 / (alpha - 1))
y_star = (k_star ** alpha) * (z_star ** (1 - alpha))
c_star = y_star - delta * k_star
i_star = delta * k_star
print("Steady-State Values:")
print("k_star:", k_star)
print("c_star:", c_star)
print("i_star:", i_star)
print("y_star:", y_star)
```

```python
# Grid for capital stock
k_min = 0.8 * k_star
k_max = 1.2 * k_star
num_points = 1000
num_firms = 300
k_grid = np.linspace(k_min, k_max, num_points)
```

```python
def calvo_price_update(p, pi, s, sigma):
    num_firms = p.shape[1]
    
    # Compute aggregate price level
    try:
        p_tilde = ((1-s)*np.exp(pi) + s*np.mean(p))**(1/(1-s))
    except OverflowError:
        p_tilde = 1e10
    
    # Update prices for firms that can adjust
    for j in range(num_firms):
        if np.random.uniform() < s:
            p[:, j] = p_tilde * np.exp(np.random.normal(0, sigma, size=(p.shape[0],)))
    
    return p
```
